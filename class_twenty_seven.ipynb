```python
### mapping
# Correction dictionary for common brand/model issues
corrections = {
    "maxda": "mazda",
    "vw": "volkswagen",
    "vokswagen": "volkswagen",
    "porcshce": "porsche",
    "toyouta": "toyota",
    "audi 100ls": "audi 100 ls",  # normalize spacing
}
```

```python
def fix_car_name(name):
    name = name.lower().strip()  # clean casing and spaces
    for wrong, correct in corrections.items():
        if name.startswith(wrong):  # match brand typos
            name = name.replace(wrong, correct, 1)
    return name
```

```python
df['CarName'] = df['CarName'].map(fix_car_name)
```

```python
x = df.drop("price",axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
```
